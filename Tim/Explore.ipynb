```python
list_of_mode = ['STEROID', 'ANTIVIRALS', 'FATIGUE', 'MALAISE', 'ANOREXIA', 'LIVER BIG',  'LIVER FIRM', 'SPLEEN PALPABLE', 'SPIDERS', 'ASCITES',	'VARICES', 'HISTOLOGY']
list_of_mean = ['BILIRUBIN', 'ALK PHOSPHATE', 'SGOT', 'ALBUMIN', 'PROTIME']
```

```python
# Convert to floats
for col in list_of_mean:
    train[col] = pd.to_numeric(train[col])

for col in ['HISTOLOGY', 'STEROID', 'ANTIVIRALS']:
    train[col] = train[col].astype('object')

# Convert to floats
for col in list_of_mean:
    test[col] = pd.to_numeric(test[col])

for col in ['HISTOLOGY', 'STEROID', 'ANTIVIRALS']:
    test[col] = test[col].astype('object')
```

```python
train.dtypes
test.dtypes
```

```python
for col in list_of_mode:
    train[col].fillna(train[col].mode().iloc[0], inplace=True)

for col in list_of_mean:
    train[col].fillna(train[col].mean(), inplace=True)

train


for col in list_of_mode:
    test[col].fillna(test[col].mode().iloc[0], inplace=True)

for col in list_of_mean:
    test[col].fillna(test[col].mean(), inplace=True)

test
```

```python
for col in train.columns:
    num = train[col].isna().sum()
    print(f" {col} : {num} ")

for col in test.columns:
    num = test[col].isna().sum()
    print(f" {col} : {num} ")
```

```python
from sklearn.preprocessing import OneHotEncoder

SEX = train[['SEX']]

OneHot = OneHotEncoder(drop='first')
OneHot.fit(SEX)
SEX_onehot = OneHot.transform(SEX).toarray()

SEX_df = pd.DataFrame(SEX_onehot, columns=OneHot.get_feature_names_out(), index=SEX.index)
SEX_df['SEX_M']
```

```python
SEX = test[['SEX']]

OneHot = OneHotEncoder(drop='first')
OneHot.fit(SEX)
SEX_onehot = OneHot.transform(SEX).toarray()

SEX_test = pd.DataFrame(SEX_onehot, columns=OneHot.get_feature_names_out(), index=SEX.index)
SEX_test['SEX_M']
```

```python
# Convert to floats
for col in list_of_mean:
    train[col] = pd.to_numeric(train[col])

for col in list_of_mode:
    train[col] = train[col].astype('object')

# Convert to floats
for col in list_of_mean:
    test[col] = pd.to_numeric(test[col])

for col in list_of_mode:
    test[col] = test[col].astype('object')
```

```python
train.dtypes
```

```python
for col in list_of_mode:
    train[col] = train[col].map(lambda x : str(x).replace("2", "0"))

for col in list_of_mode:
    test[col] = test[col].map(lambda x : str(x).replace("2", "0"))
```

```python
train.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler

numerics = train.select_dtypes(include=np.number)

scaler = MinMaxScaler()
scaled = scaler.fit_transform(numerics)

numerics_df = pd.DataFrame(scaled, columns=numerics.columns)

cats = train.drop(numerics.columns, axis=1)

merge = pd.merge(right=numerics_df, left=cats, right_index=True, left_index=True)

train = merge[[x for x in merge.columns if x != 'Class'] + ['Class']]
display(train.head(50))
```

```python
numerics = test.select_dtypes(include=np.number)

scaled = scaler.transform(numerics)

numerics_df = pd.DataFrame(scaled, columns=numerics.columns)

cats = test.drop(numerics.columns, axis=1)

merge = pd.merge(right=numerics_df, left=cats, right_index=True, left_index=True)

test = merge[[x for x in merge.columns if x != 'Class']]
test
```
